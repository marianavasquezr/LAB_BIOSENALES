a. Cree el siguiente par de vectores:

a = [3.1,1, -0.5, -3.2, 6], b= [1, 3, 2.2, 5.1, 1]



```python
import numpy as np
```

```python
a = np.array([3.1,1,-0.5,-3.2,6])
b = np.array([1,3,2.2,5.1,1])

print("Vector a: ", a)
print("Vector b: ", b)
```


b. Implemente la multiplicación escalar de a.b ¿Qué se debe realizar para poder multiplicar los dos vectores?

R/ La condicición es que los dos vectores deben tener la misma dimensión (igual número de elementos).

Aquí ambos tienen 5 elementos entonces sí se puede realizar la multiplicación de vectores

```python
producto_escalar = np.dot(a, b)
print("Producto escalar:", producto_escalar)
```

c. Implemente la multiplicación punto a punto de a.b.

```python
mult_esc = a*b
print("Multiplicación punto a punto: ", mult_esc)
```

f. Consulte la función de los siguientes comandos de la librería numpy: ones, round, ceil, floor, implemente cada función para ejemplificar su uso.

```python
# ones: Crea un arreglo lleno de unos (1) con la forma que le indiquemos.
ones = np.ones((2, 3)) 
print("\n Matriz de unos: \n", ones)

# round: Redondea los valores decimales al número de decimales indicado.
round = np.round(1.87899, 3)
print("\n Redondeado:", round)

# ceil: Redondea hacia arriba al entero más cercano.
ceil = np.ceil(matrizA)
print("\n Ceil: \n", ceil)

# floor: Redondea hacia abajo al entero más cercano.
floor = np.floor(matrizA)
print("\n Floor: \n", floor)
```
