j. Construya la función 𝑦[𝑛]=sin⁡(𝜋∗0.12𝑛) en el intervalo 0 <= n <= 100.

```python

n = np.arange(0, 101)           # Creamos el vector n desde 0 hasta 100 (101 no se incluye)
y = np.sin(np.pi * 0.12 * n)    # Construimos la señal y[n] = sin(pi * 0.12 * n)
plt.xlabel("n")                 # Etiqueta del eje x
plt.ylabel("y[n]")              # Etiqueta del eje y
plt.title("Señal y[n] = sin(pi * 0.12 n)")  # Título de la gráfica
plt.plot(n,y)                
```


k. Construya otra senoidal 𝑦2[𝑛]=cos(2𝜋∗0.03𝑛)


```python
y2 = np.cos(2*np.pi * 0.03 * n)    # Construimos la señal y2[n]
plt.xlabel("n")                 # Etiqueta del eje x
plt.ylabel("y2[n]")              # Etiqueta del eje y
plt.title("Señal y2[n] = cos(2pi * 0.03 n)")  # Título de la gráfica
plt.plot(n,y2) 
```

```python
s = np.add(y,y2)
#print("suma de señales: \n", s)

plt.xlabel("n")                 # Etiqueta del eje x
plt.ylabel("y3[n]")              # Etiqueta del eje y
plt.title("Señal s[n] = y[n]+y2[n]")  # Título de la gráfica
plt.plot(n,s) 

```

```python
t = np.multiply(y,y2)
#print("multiplicación de señales: \n", t)

plt.xlabel("n")                 # Etiqueta del eje x
plt.ylabel("y4[n]")              # Etiqueta del eje y
plt.title("Señal t[n] = y[n]*y2[n]")  # Título de la gráfica
plt.plot(n,t) 
```

1. Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
import pandas as pd

def notas_estadisticas(diccionario_notas):
    """
    Recibe un diccionario con las notas de los alumnos
    y devuelve una Serie con:
    nota mínima, máxima, media y desviación típica.
    """
    
    # Convertimos el diccionario en una Serie
    notas = pd.Series(diccionario_notas)
    
    # Calculamos estadísticas
    resultado = pd.Series({
        "Nota mínima": notas.min(),
        "Nota máxima": notas.max(),
        "Media": notas.mean(),
        "Desviación típica": notas.std()
    })
    
    return resultado


# Ejemplo
notas = {
    "Ana": 4.5,
    "Luis": 3.8,
    "Carlos": 4.2,
    "Sofía": 2.9
}

print(notas_estadisticas(notas))
```
