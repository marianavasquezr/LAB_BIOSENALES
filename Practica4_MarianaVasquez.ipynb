### **Etapas del registro de biopotenciales y componentes de un sistema de adquisición**

La adquisición de biopotenciales constituye un proceso fundamental dentro de la ingeniería biomédica, ya que permite obtener información relevante sobre el estado fisiológico de diferentes sistemas del cuerpo humano, como el sistema cardiovascular y muscular. Este proceso involucra una serie de etapas que van desde la generación de los potenciales de acción en las células excitables hasta su registro y procesamiento mediante sistemas electrónicos especializados.

### **Esquema general del proceso de adquisición**

El proceso de adquisición de biopotenciales puede describirse mediante la siguiente secuencia de etapas:

Generación del potencial de acción → Propagación en el tejido → Captación mediante electrodos → Acondicionamiento de la señal → Conversión analógica-digital → Procesamiento y visualización

Cada una de estas etapas cumple una función específica y es determinante para garantizar la calidad y confiabilidad de la señal adquirida.

### **Descripción de las etapas del registro de biopotenciales**

* **Generación del potencial de acción**

Los biopotenciales se originan en células excitables como neuronas y fibras musculares, debido a diferencias en la concentración de iones a ambos lados de la membrana celular. Estas diferencias generan un potencial eléctrico que, al alcanzar cierto umbral, produce un potencial de acción caracterizado por procesos de despolarización y repolarización (Malmivuo & Plonsey, 1995).

En el caso del electrocardiograma (ECG), la señal refleja la actividad eléctrica del corazón asociada al ciclo cardíaco, mientras que en la electromiografía (EMG) representa la activación eléctrica de las fibras musculares en respuesta a estímulos nerviosos (Webster, 2009).

* **Propagación a través del tejido biológico**

Una vez generado, el potencial de acción se propaga a través del tejido biológico mediante conducción eléctrica. Durante este proceso, la señal sufre atenuación debido a las propiedades resistivas y capacitivas de los tejidos, lo que reduce significativamente su amplitud al llegar a la superficie de la piel (Malmivuo & Plonsey, 1995).

Adicionalmente, la señal puede verse afectada por la superposición de otras fuentes bioeléctricas y ruido fisiológico, lo cual dificulta su correcta interpretación si no se implementan técnicas adecuadas de adquisición y procesamiento.

* **Captación mediante electrodos**

La captación de los biopotenciales se realiza mediante electrodos, los cuales actúan como transductores que convierten las corrientes iónicas del cuerpo en señales eléctricas medibles. En aplicaciones como ECG y EMG de superficie, se emplean electrodos colocados sobre la piel en posiciones específicas que permiten registrar diferencias de potencial entre dos puntos (Webster, 2009).

La calidad de la señal adquirida depende en gran medida de factores como la correcta ubicación de los electrodos, la preparación de la piel y la impedancia de contacto. Una alta impedancia o una mala colocación puede introducir ruido y artefactos en la señal.

* **Acondicionamiento de la señal**

Dado que los biopotenciales presentan amplitudes muy bajas (del orden de microvoltios a milivoltios), es necesario realizar un acondicionamiento de la señal antes de su digitalización. Este proceso incluye principalmente la amplificación y el filtrado.

La amplificación se lleva a cabo mediante amplificadores de instrumentación, los cuales poseen alta ganancia y un elevado rechazo de modo común (CMRR), lo que permite reducir interferencias externas como el ruido de la red eléctrica (Webster, 2009).

Por otro lado, el filtrado permite eliminar componentes no deseados de la señal. Por ejemplo, en el caso del ECG, se emplean filtros pasa-bajas y pasa-altas para limitar el rango de frecuencia de interés, así como filtros notch para eliminar el ruido de 50/60 Hz (National Library of Medicine, 2020).

* **Conversión analógica-digital (ADC)**

Una vez acondicionada, la señal analógica debe ser convertida a formato digital para su procesamiento mediante sistemas computacionales. Este proceso se realiza a través de un convertidor analógico-digital (ADC), como el que se encuentra integrado en plataformas de adquisición como Arduino.

El ADC muestrea la señal en intervalos de tiempo definidos y asigna valores discretos, permitiendo así su almacenamiento y posterior análisis (Webster, 2009).

* **Procesamiento y visualización**

Finalmente, la señal digitalizada puede ser procesada mediante software especializado para su análisis e interpretación. Este procesamiento puede incluir filtrado digital, extracción de características, análisis en el dominio del tiempo y de la frecuencia, así como visualización gráfica de la señal (Malmivuo & Plonsey, 1995).

En el caso del ECG, este análisis permite identificar parámetros como la frecuencia cardíaca y la morfología de las ondas, mientras que en el EMG permite evaluar la actividad muscular, la fuerza y la fatiga.

### **Componentes de un sistema de adquisición de biopotenciales**

Un sistema de adquisición de bioseñales está compuesto por varios elementos interconectados que permiten la correcta captura y procesamiento de la señal. Entre los principales componentes se encuentran:

* **Electrodos:** encargados de captar la señal bioeléctrica desde la superficie del cuerpo.
* **Amplificador de instrumentación:** encargado de amplificar la señal y rechazar interferencias.
* **Filtros analógicos:** utilizados para eliminar ruido y limitar el rango de frecuencia.
* **Convertidor analógico-digital (ADC):** encargado de digitalizar la señal.
* **Unidad de procesamiento:** microcontrolador o computador que procesa los datos.
* **Sistema de visualización:** software que permite observar y analizar la señal adquirida.

Estos componentes trabajan de manera conjunta para garantizar que las señales adquiridas sean representativas de la actividad fisiológica real y puedan ser utilizadas con fines diagnósticos, investigativos o de control (Webster, 2009).

# **Fuentes de ruido en el registro de biopotenciales y estrategias de mitigación**

La adquisición de biopotenciales como el ECG y el EMG está afectada por múltiples fuentes de ruido que pueden degradar la calidad de la señal y dificultar su interpretación. Estas interferencias pueden originarse en diferentes etapas del sistema de adquisición, incluyendo la interfaz electrodo-piel, el entorno y los componentes electrónicos del sistema. Estudios recientes destacan que los artefactos pueden generarse en distintas partes de la cadena de adquisición, como los electrodos, los cables y el sistema de amplificación . [5]

Las fuentes de ruido pueden clasificarse en estáticas (relativamente constantes en el tiempo) y variantes en el tiempo (dinámicas), lo cual es clave para definir estrategias adecuadas de mitigación.



### **Análisis de las fuentes de ruido**

Los estudios actuales muestran que una de las principales fuentes de interferencia en bioseñales son los artefactos por movimiento, los cuales pueden originarse por desplazamientos del cuerpo, respiración o movimiento de los electrodos, generando variaciones en la línea base y dificultando la detección de eventos como el complejo QRS en ECG .

Adicionalmente, se ha demostrado que estos artefactos están estrechamente relacionados con cambios en la impedancia electrodo-piel, ya que el movimiento altera el equilibrio electroquímico en la interfaz, introduciendo señales espurias en la medición .

Por otra parte, investigaciones recientes en EMG indican que las señales pueden estar contaminadas por diferentes tipos de ruido, incluyendo interferencia de la red eléctrica, ruido impulsivo y contaminación cruzada con ECG, lo cual afecta el rendimiento de los sistemas de análisis si no se identifican adecuadamente .

### **Clasificación: ruido estático vs. variante en el tiempo**
Ruido estático:
Se caracteriza por ser constante o predecible, como el ruido térmico o la interferencia de la red eléctrica. Este tipo de ruido puede ser eliminado eficazmente mediante técnicas de filtrado.
Ruido variante en el tiempo:
Es dinámico e impredecible, como los artefactos por movimiento o la interferencia electromagnética. Este tipo de ruido es más complejo de tratar, ya que requiere tanto estrategias de adquisición como de procesamiento.
Estrategias de mitigación

Las estrategias para reducir el ruido en bioseñales se implementan en diferentes niveles:

A nivel de adquisición
Preparación adecuada de la piel (reducción de impedancia)
Correcta colocación de electrodos
Minimización del movimiento del paciente
Uso de cables blindados
A nivel electrónico
Amplificadores diferenciales con alto CMRR
Sistemas de puesta a tierra adecuados
Diseño de circuitos de bajo ruido
A nivel de procesamiento
Filtros notch (50/60 Hz)
Filtros pasa-banda
Técnicas avanzadas (wavelets, filtros adaptativos)

Actualmente, también se están desarrollando métodos más avanzados como algoritmos de reducción adaptativa y técnicas de aprendizaje automático para identificar y eliminar ruido en tiempo real
