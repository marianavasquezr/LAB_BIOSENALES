### **Etapas del registro de biopotenciales y componentes de un sistema de adquisición**

La adquisición de biopotenciales constituye un proceso fundamental dentro de la ingeniería biomédica, ya que permite obtener información relevante sobre el estado fisiológico de diferentes sistemas del cuerpo humano, como el sistema cardiovascular y muscular. Este proceso involucra una serie de etapas que van desde la generación de los potenciales de acción en las células excitables hasta su registro y procesamiento mediante sistemas electrónicos especializados.

### **Esquema general del proceso de adquisición**

El proceso de adquisición de biopotenciales puede describirse mediante la siguiente secuencia de etapas:

Generación del potencial de acción → Propagación en el tejido → Captación mediante electrodos → Acondicionamiento de la señal → Conversión analógica-digital → Procesamiento y visualización

Cada una de estas etapas cumple una función específica y es determinante para garantizar la calidad y confiabilidad de la señal adquirida.

### **Descripción de las etapas del registro de biopotenciales**

* **Generación del potencial de acción**

Los biopotenciales se originan en células excitables como neuronas y fibras musculares, debido a diferencias en la concentración de iones a ambos lados de la membrana celular. Estas diferencias generan un potencial eléctrico que, al alcanzar cierto umbral, produce un potencial de acción caracterizado por procesos de despolarización y repolarización [1].

En el caso del electrocardiograma (ECG), la señal refleja la actividad eléctrica del corazón asociada al ciclo cardíaco, mientras que en la electromiografía (EMG) representa la activación eléctrica de las fibras musculares en respuesta a estímulos nerviosos [2].

* **Propagación a través del tejido biológico**

Una vez generado, el potencial de acción se propaga a través del tejido biológico mediante conducción eléctrica. Durante este proceso, la señal sufre atenuación debido a las propiedades resistivas y capacitivas de los tejidos, lo que reduce significativamente su amplitud al llegar a la superficie de la piel [1].

Adicionalmente, la señal puede verse afectada por la superposición de otras fuentes bioeléctricas y ruido fisiológico, lo cual dificulta su correcta interpretación si no se implementan técnicas adecuadas de adquisición y procesamiento.

* **Captación mediante electrodos**

La captación de los biopotenciales se realiza mediante electrodos, los cuales actúan como transductores que convierten las corrientes iónicas del cuerpo en señales eléctricas medibles. En aplicaciones como ECG y EMG de superficie, se emplean electrodos colocados sobre la piel en posiciones específicas que permiten registrar diferencias de potencial entre dos puntos [2].

La calidad de la señal adquirida depende en gran medida de factores como la correcta ubicación de los electrodos, la preparación de la piel y la impedancia de contacto. Una alta impedancia o una mala colocación puede introducir ruido y artefactos en la señal.

* **Acondicionamiento de la señal**

Dado que los biopotenciales presentan amplitudes muy bajas (del orden de microvoltios a milivoltios), es necesario realizar un acondicionamiento de la señal antes de su digitalización. Este proceso incluye principalmente la amplificación y el filtrado.

La amplificación se lleva a cabo mediante amplificadores de instrumentación, los cuales poseen alta ganancia y un elevado rechazo de modo común (CMRR), lo que permite reducir interferencias externas como el ruido de la red eléctrica [2].

Por otro lado, el filtrado permite eliminar componentes no deseados de la señal. Por ejemplo, en el caso del ECG, se emplean filtros pasa-bajas y pasa-altas para limitar el rango de frecuencia de interés, así como filtros notch para eliminar el ruido de 50/60 Hz [3].

* **Conversión analógica-digital (ADC)**

Una vez acondicionada, la señal analógica debe ser convertida a formato digital para su procesamiento mediante sistemas computacionales. Este proceso se realiza a través de un convertidor analógico-digital (ADC), como el que se encuentra integrado en plataformas de adquisición como Arduino.

El ADC muestrea la señal en intervalos de tiempo definidos y asigna valores discretos, permitiendo así su almacenamiento y posterior análisis [2].

* **Procesamiento y visualización**

Finalmente, la señal digitalizada puede ser procesada mediante software especializado para su análisis e interpretación. Este procesamiento puede incluir filtrado digital, extracción de características, análisis en el dominio del tiempo y de la frecuencia, así como visualización gráfica de la señal [1].

En el caso del ECG, este análisis permite identificar parámetros como la frecuencia cardíaca y la morfología de las ondas, mientras que en el EMG permite evaluar la actividad muscular, la fuerza y la fatiga.

### **Componentes de un sistema de adquisición de biopotenciales**

Un sistema de adquisición de bioseñales está compuesto por varios elementos interconectados que permiten la correcta captura y procesamiento de la señal. Entre los principales componentes se encuentran:

* **Electrodos:** encargados de captar la señal bioeléctrica desde la superficie del cuerpo.
* **Amplificador de instrumentación:** encargado de amplificar la señal y rechazar interferencias.
* **Filtros analógicos:** utilizados para eliminar ruido y limitar el rango de frecuencia.
* **Convertidor analógico-digital (ADC):** encargado de digitalizar la señal.
* **Unidad de procesamiento:** microcontrolador o computador que procesa los datos.
* **Sistema de visualización:** software que permite observar y analizar la señal adquirida.

Estos componentes trabajan de manera conjunta para garantizar que las señales adquiridas sean representativas de la actividad fisiológica real y puedan ser utilizadas con fines diagnósticos, investigativos o de control [2].

# **Fuentes de ruido en el registro de biopotenciales y estrategias de mitigación**

La adquisición de biopotenciales como el ECG y el EMG está afectada por múltiples fuentes de ruido que pueden degradar la calidad de la señal y dificultar su interpretación. Estas interferencias pueden originarse en diferentes etapas del sistema de adquisición, incluyendo la interfaz electrodo-piel, el entorno y los componentes electrónicos del sistema. Estudios recientes destacan que los artefactos pueden generarse en distintas partes de la cadena de adquisición, como los electrodos, los cables y el sistema de amplificación [5].

Las fuentes de ruido pueden clasificarse en estáticas (relativamente constantes en el tiempo) y variantes en el tiempo (dinámicas), lo cual es clave para definir estrategias adecuadas de mitigación.



### **Análisis de las fuentes de ruido**

Los estudios actuales muestran que una de las principales fuentes de interferencia en bioseñales son los artefactos por movimiento, los cuales pueden originarse por desplazamientos del cuerpo, respiración o movimiento de los electrodos, generando variaciones en la línea base y dificultando la detección de eventos como el complejo QRS en ECG [6]. 

Adicionalmente, se ha demostrado que estos artefactos están estrechamente relacionados con cambios en la impedancia electrodo-piel, ya que el movimiento altera el equilibrio electroquímico en la interfaz, introduciendo señales espurias en la medición [7].

Por otra parte, investigaciones recientes en EMG indican que las señales pueden estar contaminadas por diferentes tipos de ruido, incluyendo interferencia de la red eléctrica, ruido impulsivo y contaminación cruzada con ECG, lo cual afecta el rendimiento de los sistemas de análisis si no se identifican adecuadamente [8].

### **Clasificación: ruido estático vs. variante en el tiempo**

**Ruido estático:**
Se caracteriza por ser constante o predecible, como el ruido térmico o la interferencia de la red eléctrica. Este tipo de ruido puede ser eliminado eficazmente mediante técnicas de filtrado.

**Ruido variante en el tiempo:**
Es dinámico e impredecible, como los artefactos por movimiento o la interferencia electromagnética. Este tipo de ruido es más complejo de tratar, ya que requiere tanto estrategias de adquisición como de procesamiento.
Estrategias de mitigación

Las estrategias para reducir el ruido en bioseñales se implementan en diferentes niveles:

- A nivel de adquisición
    - Preparación adecuada de la piel (reducción de impedancia)
    - Correcta colocación de electrodos
    - Minimización del movimiento del paciente
    - Uso de cables blindados
- A nivel electrónico
    - Amplificadores diferenciales con alto CMRR
    - Sistemas de puesta a tierra adecuados
    - Diseño de circuitos de bajo ruido
- A nivel de procesamiento
    - Filtros notch (50/60 Hz)
    - Filtros pasa-banda
    - Técnicas avanzadas (wavelets, filtros adaptativos)

Actualmente, también se están desarrollando métodos más avanzados como algoritmos de reducción adaptativa y técnicas de aprendizaje automático para identificar y eliminar ruido en tiempo real [5].

# **Problemática del acoplamiento piel–electrodo y estrategias de mejora**

El acoplamiento piel–electrodo constituye una de las etapas más críticas en la adquisición de biopotenciales, ya que representa la interfaz directa entre el sistema biológico y el sistema electrónico de medición. La calidad de esta interfaz determina en gran medida la fidelidad de la señal registrada, especialmente en técnicas no invasivas como el electrocardiograma (ECG), la electromiografía de superficie (sEMG) y la electroencefalografía (EEG).

**Problemática del acoplamiento piel–electrodo**

La principal problemática del acoplamiento piel–electrodo radica en que la piel no es un medio conductor ideal. La capa más externa, el estrato córneo, presenta una alta resistencia eléctrica, lo que genera una impedancia elevada y variable en la interfaz electrodo–piel. Esta impedancia puede introducir distorsiones, atenuación de la señal y susceptibilidad a interferencias externas [9].

Además, en esta interfaz se producen fenómenos electroquímicos que generan un potencial de media celda, el cual puede provocar una deriva en la línea base de la señal. Este fenómeno se ve agravado por factores como la sudoración, la temperatura y el movimiento del paciente, los cuales alteran continuamente las condiciones del contacto [10].

Otra fuente importante de error son los artefactos por movimiento, que se originan cuando hay desplazamientos relativos entre el electrodo y la piel. Estos artefactos generan variaciones abruptas en la señal, dificultando la identificación de características relevantes como el complejo QRS en ECG o los patrones de activación en EMG [11].

Asimismo, una mala adherencia del electrodo o una preparación inadecuada de la piel puede generar una baja relación señal-ruido (SNR), lo que compromete la calidad del registro y puede conducir a interpretaciones erróneas.

**Estrategias para mejorar el acoplamiento piel–electrodo**

Para mitigar los problemas asociados al acoplamiento piel–electrodo, se han desarrollado diversas estrategias tanto a nivel de preparación del paciente como de diseño de los electrodos:

- **Preparación de la piel**
Limpieza con alcohol para eliminar grasa, sudor y suciedad
En algunos casos, abrasión ligera para reducir la impedancia del estrato córneo

Estas acciones permiten disminuir significativamente la impedancia de contacto y mejorar la conductividad [9].

- **Uso de geles conductores**

El uso de gel electrolítico facilita la conducción eléctrica entre la piel y el electrodo, reduciendo la impedancia y estabilizando la señal. Este tipo de gel permite una mejor adaptación a las irregularidades de la piel [10].

- **Selección adecuada de electrodos**
Electrodos de Ag/AgCl (cloruro de plata) son los más utilizados debido a su estabilidad electroquímica
Electrodos secos o textiles en aplicaciones portátiles, aunque con mayor susceptibilidad al ruido

Los materiales y el diseño del electrodo influyen directamente en la calidad del contacto [11].

- **Fijación mecánica adecuada**

El uso de adhesivos o bandas elásticas permite mantener el electrodo en posición estable, reduciendo los artefactos por movimiento.

- **Reducción del movimiento del paciente**

Especialmente importante en ECG y EEG, donde el movimiento puede introducir ruido significativo.

- **Uso de amplificadores con alto CMRR**

Permiten reducir el impacto de variaciones en la impedancia y mejorar la calidad de la señal registrada.

- **Importancia crítica del acoplamiento piel–electrodo**

El acoplamiento piel–electrodo es considerado la etapa más crítica en el registro de señales de superficie debido a que constituye el primer punto de contacto en la cadena de adquisición. Cualquier error o deficiencia en esta interfaz no puede ser completamente corregido en etapas posteriores del procesamiento.

En particular:

Determina la relación señal-ruido (SNR) desde el inicio
Introduce ruido que puede ser indistinguible de la señal biológica
Afecta la estabilidad de la línea base
Puede generar artefactos que simulan eventos fisiológicos reales

Estudios recientes destacan que la calidad del contacto electrodo–piel es uno de los factores más determinantes en la precisión de sistemas de monitoreo biomédico, especialmente en aplicaciones portátiles y de largo plazo [11].

En consecuencia, un acoplamiento deficiente puede invalidar completamente la medición, independientemente de la calidad del resto del sistema de adquisición.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import neurokit2 as nk
from ecgdetectors import Detectors

# 1. Cargar señal desde .txt
data = np.loadtxt("DATOS_ECG.txt")

# Frecuencia de muestreo
fs = 100  # Se asumio teniendo en cuenta que son 10 picos R para 1000 muestras.

# 2. Procesamiento con NeuroKit
# nk.ecg_process devuelve un diccionario con la señal procesada y la información relevante, incluyendo los picos R detectados.
signals, info = nk.ecg_process(data, sampling_rate=fs)
# Extraemos los índices de los picos R detectados por NeuroKit
rpeaks_nk = info["ECG_R_Peaks"] 

# 3. Procesamiento con Detectors
# Detectors es una biblioteca que implementa varios algoritmos de detección de picos R. 
detectors = Detectors(fs)
# El método pan_tompkins_detector devuelve los índices de los picos R detectados.
rpeaks_det = detectors.pan_tompkins_detector(data)

# 4. Comparación visual
# Se crea una figura para mostrar la señal ECG y los picos R detectados por ambos métodos.
plt.figure(figsize=(12,5)) 
plt.plot(data, label="ECG")

# Se grafican los picos R detectados por NeuroKit en rojo y los detectados por Detectors en verde.
plt.scatter(rpeaks_nk, data[rpeaks_nk], color='red', label="NeuroKit", s=20)
plt.scatter(rpeaks_det, data[rpeaks_det], color='green', label="Detectors", s=20)

plt.title("Detección de picos R")
plt.legend()
plt.xlabel("Muestras")
plt.ylabel("Amplitud")
plt.show()
```

### **Análisis de resultados**

A partir de la comparación visual, se evidencia que el algoritmo de NeuroKit presenta una detección más precisa y consistente de los picos R, alineándose adecuadamente con los máximos del complejo QRS. En contraste, el algoritmo Detectors muestra errores sistemáticos al identificar puntos que no corresponden a latidos reales, lo que lo hace inadecuado para el análisis de intervalos RR y, por ende, para el cálculo de la variabilidad de la frecuencia cardíaca. Por esta razón, se selecciona NeuroKit como método de referencia para el procesamiento de la señal.

```python
# 5. Cálculo de intervalos RR
rr_intervals = np.diff(rpeaks_nk) / fs #en segundos
rr_ms = rr_intervals * 1000  # en milisegundos
print("Intervalos RR (ms):", rr_ms)
```

### **Librería hrvanalysis y función get_time_domain_features**

La **librería hrvanalysis** es una herramienta de código abierto en Python diseñada para el análisis de la variabilidad de la frecuencia cardíaca (HRV) a partir de intervalos NN (normal-to-normal). Esta librería permite calcular métricas tanto en el dominio del tiempo como en el dominio de la frecuencia, así como realizar preprocesamiento de la señal, incluyendo eliminación de artefactos y detección de latidos.

Según su documentación oficial, hrvanalysis proporciona un conjunto de funciones enfocadas en facilitar el análisis fisiológico del sistema cardiovascular, siendo ampliamente utilizada en estudios de bioingeniería, medicina y procesamiento de bioseñales [12].

La función **get_time_domain_features** permite extraer automáticamente las principales características de HRV en el dominio del tiempo a partir de una serie de intervalos RR expresados en milisegundos.

De acuerdo con la documentación oficial, esta función recibe como entrada una lista de intervalos NN y devuelve un diccionario con múltiples métricas estadísticas relevantes para el análisis de la variabilidad cardíaca [12].

**Principales métricas calculadas**

Entre las variables más importantes que calcula la función se encuentran:

* **Mean NN (mean_nni):** promedio de los intervalos RR, relacionado con la frecuencia cardíaca media.
* **SDNN:** desviación estándar de los intervalos NN, que refleja la variabilidad global del sistema cardiovascular.
* **RMSSD:** raíz cuadrática media de las diferencias sucesivas entre intervalos NN, asociada principalmente con la actividad parasimpática del sistema nervioso autónomo.
* **NN50:** número de pares de intervalos consecutivos que difieren en más de 50 ms.
* **pNN50:** porcentaje de dichos intervalos respecto al total, indicador de variabilidad a corto plazo.

Estas métricas son ampliamente utilizadas en la evaluación del equilibrio simpático-parasimpático y en el estudio de la regulación autonómica del corazón [13].

```python
from hrvanalysis import get_time_domain_features

# Calcular características HRV
caract_rr = get_time_domain_features(rr_ms) #Es un diccionario con las características calculadas
# Mostrar resultados
for key, value in caract_rr.items():
    print(f"{key}: {value}")
```
